```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Dell\Desktop\GOLD.csv")
data
```

```python
data['Date'] = data['Date'].str.strip()  
data['date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y') 
data = data.dropna() 
data['Gold Price JD'] = pd.to_numeric(data['Gold Price JD'], errors='coerce') 
data[' Gold Price USD'] = pd.to_numeric(data[' Gold Price USD'], errors='coerce') 
data['geopolitical_situation_numeric'] = data['Geopolitical situation'].map({'Stable': 0, 'Tense': 1})
print(data.info())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(data[' Gold Price USD'], data['Gold Price JD'])
plt.title('Gold Price JD vs  Gold Price USD')
plt.xlabel(' Gold Price USD')
plt.ylabel('Gold Price JD')

# Set limits for better visualization
plt.xlim(50, 60)  # Adjust the limits according to your data range
plt.ylim(30, 50)  # Adjust the limits according to your data range

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import numpy as np 
import matplotlib.pyplot as plt
#  Number of simulations 
num_simulations = 5
# List to store the forecasts
forecasts = []
for i in range(num_simulations):
    # Build the ARIMA model   
    model_arima = ARIMA(data['Gold Price JD'], exog=data[[' Gold Price USD', 'geopolitical_situation_numeric']], order=(5, 1, 0))
    model_fit = model_arima.fit()
    # Prepare exogenous variables for future predictions  
    exog_future = data[[' Gold Price USD', 'geopolitical_situation_numeric']].iloc[-1:]
    exog_future = pd.concat([exog_future] * 10, ignore_index=True)  # Repeat the last value 10 times
    # Optionally add some random noise to exogenous variables to get slightly different predictions    
    exog_future += np.random.normal(0, 0.01, exog_future.shape)
    #Gaussian distribution mean = 0,stander divitain = 0.01 

    # Forecast the future values
    forecast = model_fit.forecast(steps=10, exog=exog_future)   
    forecasts.append(forecast)
    # Prepare dates for future predictions
    future_dates = pd.date_range(start=data['date'].iloc[-1], periods=11, freq='D')[1:]
    # Create a DataFrame for the forecasted data 
    forecast_df = pd.DataFrame({'date': future_dates, 'forecast': forecast})
    # Plotting
    plt.figure(figsize=(12, 6))   
    plt.plot(data['date'], data['Gold Price JD'], label='Actual Prices')
    plt.plot(forecast_df['date'], forecast_df['forecast'], label='Forecasted Prices', linestyle='--') 
    plt.title(f'Gold Price Prediction in Jordan (simulation {i+1})')  
    plt.xlabel('Date')  
    plt.ylabel('Gold Price JD')
    plt.legend()    
    plt.show()
```
